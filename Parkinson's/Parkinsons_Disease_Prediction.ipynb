# ***Parkinson's Disease Prediction***

### Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
```

### Importing the dataset

```python
data = pd.read_csv('parkinsons.data')
data.head()
```

### Grouping the data based on the target variable

```python
data.groupby('status').mean(numeric_only=True)
```

### Data Pre-processing

```python
X = data.drop(columns=['name','status'],axis=1)
Y = data['status']
```

```python
X
```

```python
Y
```

### Splitting the dataset into training and test data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

### Data Standardization

```python
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

### Model training using SVM

```python
svm = SVC(kernel='rbf')
svm.fit(X_train,Y_train)
```

### Model Evaluation

```python
# Accuracy Score of training data
X_train_pred = svm.predict(X_train)
training_accuracy = accuracy_score(Y_train,X_train_pred)
print('Accuracy of the training data:',training_accuracy*100)
```

```python
# Accuracy Score of test data
X_test_pred = svm.predict(X_test)
test_accuracy = accuracy_score(Y_test,X_test_pred)
print('Accuracy of the test data:',test_accuracy*100)
```

### Building a ***Predictive System***

```python
input_data = [237.22600,247.32600,225.22700,0.00298,0.00001,0.00169,0.00182,0.00507,0.01752,0.16400,0.01035,0.01024,0.01133,0.03104,0.00740,22.73600,0.305062,0.654172,-7.310550,0.098648,2.416838,0.095032
]

# converting input data to array
input_arr = np.asarray(input_data)

# reshaping the array
input_reshaped_arr = input_arr.reshape(1,-1)

# standardising the data
std_data = scaler.transform(input_reshaped_arr)

prediction = svm.predict(std_data)

if (prediction[0] == 0):
  print('The person is healthy!!')
else:
  print("The person is parkinson's positive")
```
